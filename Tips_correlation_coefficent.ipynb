```python
#importing required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Ignore harmless warnings
import warnings
warnings.filterwarnings('ignore')#IMPORT THE PENGUIN DATA SET
data = sns.load_dataset('tips')
# Data dimensions
print(f' The data has {data.shape[0]} rows and {data.shape[1]} columns')



round(data.isnull().mean()*100,2)
```

```python
#data sample
data.head()
```

```python
#Use tips data and find out correlation for all numerical values
data.corr(numeric_only='True')
```

```python
#Use tips data and find out correlation for all numerical values
sns.pairplot(data)
plt.show()
```

```python
#Try combination of numerical values with respect to hue
sns.pairplot(data=data,
             x_vars=['total_bill','tip','size'],
             y_vars=['tip'],
             hue='sex',
             height=5, aspect=1, kind='scatter')     # reg - regression
plt.show()
```
